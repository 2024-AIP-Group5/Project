```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './result/'
epoch = 300
data_usage = 1.0 
modelist = ['probe4', 'final', 'final2']


loss_dict = {}
for mode in modelist:
    loss_dict[mode+'_train'] = np.loadtxt(f'{path}trainLoss_{mode}_epoch_{epoch}_dataUsage_{data_usage}')
    # loss_dict[mode+'_val'] = np.loadtxt(f'{path}valLoss_{mode}_epoch_{epoch}_dataUsage_{data_usage}')

    
for label, losses in loss_dict.items():
    plt.plot(losses, label=label)


plt.title("loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
# plt.ylim(1.3e4, 2.8e4)
# plt.ylim(1.5e4, 1.8e4)
plt.ylim(1.3e4, 1.6e4)
plt.show()
```
